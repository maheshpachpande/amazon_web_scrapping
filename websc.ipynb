```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
searchstring="iphone"
url=f"https://www.amazon.in/s?k={searchstring}&page=2"
```

```python
driver = webdriver.Chrome()
```

```python
driver.get(url)
```

```python
html_data=BeautifulSoup(driver.page_source,'html.parser')
```

```python
print(html_data)
```

```python
no_of_pages=html_data.find('span',{'class':'s-pagination-item s-pagination-disabled'}).text
no_of_pages=int(no_of_pages)
no_of_pages
```

```python
products=html_data.find_all('div',{'data-component-type':'s-search-result'})
```

```python
from tqdm import tqdm
```

```python
title=[]
image_link=[]
price=[]
original_price=[]
ratings=[]

searchstring='iphone'

for i in tqdm(range(no_of_pages)):

    url=f"https://www.amazon.in/s?k={searchstring}&page={i+1}"
    driver.get(url)
    time.sleep(5)
    html_data=BeautifulSoup(driver.page_source,'html.parser')
    products=html_data.find_all('div',{'data-component-type':'s-search-result'})
    for product in products:
        try:
            title1=product.find('span',{'class':'a-size-medium a-color-base a-text-normal'}).text
            title.append(title1)
        except:
            title1.append('No Title Available')
        try:
            price1=product.find('span',{'class':'a-price-whole'}).text
            price.append(price1)
        except:
            price.append('No Price Available')
        try:
            original_price1=product.find('span',{'aria-hidden':'true'}).text
            original_price.append(original_price1)
        except:
            original_price.append('No Original Price Available')
        try:
            image_link1=product.find('img')['src']
            image_link.append(image_link1)
        except:
            image_link.append('No Image Available')
        try:
            ratings1=product.find('span',{'class':'a-icon-alt'}).text
            ratings.append(ratings1)
        except:
            ratings.append('No Ratings Available')
```

```python
data = pd.DataFrame({'Title':title, 'Price':price, 'Original Price':original_price, 'Ratings':ratings, 'Image links':image_link})
data
```

```python
data.to_csv('Iphone.csv')
```
